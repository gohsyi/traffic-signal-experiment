```python
from collect import *
import os

smoothing_weight = 0.25
save_path = 'log/portion.jpg'
train_curve = []
eval_block = []

eval_portion = [[], [], [], [], []]

results = ['log/4normal/results.txt', 
           'log/10normal/results.txt',
           'log/20normal/results.txt', 
           'log/50normal/results.txt',
           'log/100normal/results.txt']

labels = ['4-d signals', '10-d signals', '20-d signals', '50-d signals', '100-d signals']

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']

for no, file_name in enumerate(results):
    with open(file_name, 'r', encoding='utf8') as f:
        print('processing {} ...'.format(file_name))
        count = 0
        
        for line in f:
            if 'Train' in line:
                if count:
                    eval_data = read_eval_block(eval_block)
                    if eval_data:
                        portion = 0
                        ep, sig_row, sig_col, ac_row, ac_col = eval_data
                        print(ep)
                        for i in range(len(ac_row)):
                            if ac_row[i] == 0 and ac_col[i] == 0: portion += 1
                        portion /= len(ac_row)
                        if len(eval_portion[no]) == 0:
                            eval_portion[no].append(portion)
                        else:
                            eval_portion[no].append(portion * smoothing_weight + 
                                                    (1 - smoothing_weight) * eval_portion[no][-1])
                            if len(eval_portion[no]) >= 30: break
                    count = 0
                    eval_block = []
                    
            elif 'Eval' in line:
                eval_block.append(line)
                count += 1
    plt.plot(range(len(eval_portion[no])), np.array(eval_portion[no]), c=colors[no], label=labels[no])

print('saveing to {}'.format(save_path))

plt.legend(loc='upper right')
plt.savefig(save_path)
plt.cla()
plt.close()

print('end.')
```
